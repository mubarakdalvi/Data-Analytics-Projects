getting values as per zones for company X


```python
def impute_row(row):
  if row['Zone By Comapany X'] == 'a':
    return 0.50
  elif row['Zone By Comapany X'] == 'b':
    return 1.00
  elif row['Zone By Comapany X'] == 'c':
    return 1.25
  elif row['Zone By Comapany X'] == 'd':
    return 1.50
  elif row['Zone By Comapany X'] == 'e':
    return 2.00
zone_order['Weight slab (kg)'] = zone_order.apply(impute_row,axis=1)
```

getting values as per zones for courier company

```python
def wscourier(row):
  if row['Zone By Courier'] == 'a':
    return 0.50
  elif row['Zone By Courier'] == 'b':
    return 1.00
  elif row['Zone By Courier'] == 'c':
    return 1.25
  elif row['Zone By Courier'] == 'd':
    return 1.50
  elif row['Zone By Courier'] == 'e':
    return 2.00

zone_order['weight slab by Courier'] = zone_order.apply(wscourier,axis=1)
```

```python
zone_order.head()
```

calculating charges as per in simple we will calculate on the basis of weight and slab by zones and if weight is more than zone we will add addional charges in case of rto will add addtional rto charges as fixed rto charges will be add , because it is fixed charges

```python
def calculate_charge(row):
    Zone = ['a', 'b', 'c', 'd', 'e']
    WeightSlabs = [0.50, 1.00, 1.25, 1.50, 2.00]
    ForwardFixedCharge = [29.5, 33.0, 40.1, 45.4, 56.6]
    ForwardAdditionalWeightSlabCharge = [23.6, 28.3, 38.9, 44.8, 55.5]
    RTOFixedCharge = [13.6, 20.5, 31.9, 41.3, 50.7]
    RTOAdditionalWeightSlabCharge = [23.6, 28.3, 38.9, 44.8, 55.5]

    zone_index = Zone.index(row['Zone By Comapany X'])
    fixed_weight_slab = WeightSlabs[zone_index]
    remaining_weight = max(0, row['Weight By Company X'] - fixed_weight_slab)

    charge = ForwardFixedCharge[zone_index]
    if remaining_weight > 0:
        charge += remaining_weight * ForwardAdditionalWeightSlabCharge[zone_index]

    if row['Weight By Company X']:
        charge += RTOFixedCharge[zone_index]
        if remaining_weight > 0:
            charge += remaining_weight * RTOAdditionalWeightSlabCharge[zone_index]

    return charge

zone_order['Expected By Company X'] = zone_order.apply(calculate_charge, axis=1)
```
